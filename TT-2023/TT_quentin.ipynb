```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

Tout d'abord, nous importons nos données que nous avons déjà modifié dans le fichier csv "result\-vote". Nous modifions le type de la colonne "Code du département" en 'string', car certains départements ont un code composé de lettres. Nous mettons chaque code en 'string' afin de simplifier nos futures fonctions.


```python
import pandas as pd
dfq = pd.read_csv('result-vote.csv',sep=';',encoding='latin1',dtype={"Code du département" : str})
dfq
```

Nous créons une fonction 'stats\-dep' qui prend en argument un 'string' d, qui est un code de département. Cette fonction va renvoyer une diagramme en camembert avec les résultats en pourcentage de chaque candidats dans un département d donné en argument. Voici cette fonction :



```python
def stats_dep(d):
    df2q=dfq[(dfq['Code du département']==d)]
    dataq={'Nom':[],'Voix':[]}
    dataq['Nom'].append(df2q['Nom'][df2q.index[0]])
    dataq['Voix'].append(df2q['Voix'].sum())
    for i in range(1,12):
        dataq['Nom'].append(df2q['Nom.'+str(i)][df2q.index[0]])
        dataq['Voix'].append(df2q['Voix.'+str(i)].sum())
    df3q = pd.DataFrame(dataq)
    df3q=df3q.sort_values('Voix',ascending=False)
    palette = cm.Set3
    plt.pie(df3q['Voix'],labels=df3q['Nom'],colors=palette.colors,autopct='%1.2f%%',pctdistance=0.85)
    plt.title("Résultat pour le département "+df2q['Libell  du département'][df2q.index[0]])
    plt.show()
    return df3q
    
#Testons notre fonction pour le département du Rhône :
Ga=stats_dep('69')
```

Ayant ce jeu de données complet, nous pouvons imaginer un autre résultat lors du premier tour avec un système d'élection différent. Nous avons décidé de nous inspirer du système de vote américain. En effet, chaque département sera comme un état, et le nombre de grands électeurs par département sera déterminé par le nombre de circonscriptions que possède chaque département. Par exemple, le département du Rhône sera représenté par 14 grands électeurs et le département de Paris par 18 grands électeurs. Ainsi, nous avons décidé que le candidat arrivant en tête au sein d'un département remportera un nombre de voix égal au nombre de circonscriptions qui composent ce département. La France possédant 577 circonscriptions, 577 voix sont à partager entre les candidats. 

Pour réaliser cela, nous écrivons tout d'abord une fonction qui renvoie le gagnant au sein d'un département d passé en argument :



```python
#fonction pour avoir le gagnant par département 
def ga_dep(d):
    df2=dfq[(dfq['Code du département']==d)]
    data={'Nom':[],'Voix':[]}
    data['Nom'].append(df2['Nom'][df2.index[0]])
    data['Voix'].append(df2['Voix'].sum())
    for i in range(1,12):
        data['Nom'].append(df2['Nom.'+str(i)][df2.index[0]])
        data['Voix'].append(df2['Voix.'+str(i)].sum())
    df3 = pd.DataFrame(data)
    df3=df3.sort_values('Voix',ascending=False)
    return df3

#Regardons le gagnant dans le département du Rhône :
Gag=ga_dep('69')
Gag['Nom'][Gag.index[0]]
```

Avec cette fonction, nous allons pouvoir écrire le code qui va nous donner les résultats des votes style "américains". Nous décidons qu'il y a seulement 3 candidats qui participent à ces élections \(puisque seuls ces 3 candidats sont arrivés en tête d'un département\) : Macron, Le Pen et Mélenchon.



```python
from numpy import unique
codes = dfq['Code du département']
dataq={'Nom':[],'Voix':[]}
for i in unique(codes):
    df2q = dfq[codes == i] #permet de parcourir une seule fois le département 'i'
    Nb_circ=max(df2q['Code de la circonscription']) #nombre de circonscription pour le département 'i'
    df3q=ga_dep(i) #permet d'obtenir le gagnant pour le département 'i'
    dataq['Nom'].append(df3q['Nom'][df3q.index[0]])
    dataq['Voix'].append(Nb_circ)
tableau_liste = pd.DataFrame(dataq).values.tolist() #nous travaillons avec des tableaux pour que ce soit plus simple à manier avec une boucle for
res = [['MACRON',0],['LE PEN',0], ['M LENCHON',0]] #nos résultats
for i in range (len(tableau_liste)):
    if (tableau_liste[i][0]=='MACRON'):
             res[0][1]=res[0][1]+tableau_liste[i][1]
    elif (tableau_liste[i][0]=='LE PEN'):
             res[1][1]=res[1][1]+tableau_liste[i][1]
    elif (tableau_liste[i][0]=='M LENCHON'):
             res[2][1]=res[2][1]+tableau_liste[i][1]
Résultats=pd.DataFrame(res, columns = ['Candidat', 'Nombre de voix'])
print("Résultats du premier tour de l'élection présidentielle avec un système de vote américain :")
Résultats
```

Le résultat est différent du résultat "réel". En effet, avec ce nouveau système, E.Macron aurait récolté 52.51% des voix \(contre 27,85%\) en réalité et aurait été élu avec plus de la moitié des voix. M.Le Pen et JL.Mélenchon aurait eu respectivement 34.49% et 13% des voix \(contre 23,15% et 21,95% des voix dans la réalité\).

Nous pouvons discuter de ce système de vote. Tout d'abord, le système français permet de prendre en compte la majorité à l'échelle nationale, sans réelle sélection en amont. Le système de vote américain fait entrer en jeu une première sélection à l'échelle départementale \(ou à l'échelle d'un état\). Ce système peut être injuste pour les petits candidats, puisqu'on peut faire l'unanimité à l'échelle nationale mais ne pas arriver en tête dans un seul département \(peu probable mais mathématiquement possible\). Ainsi, on peut dire que le système de vote américain ne permet pas de refléter l'avis national, mais plutôt l'avis département par département.
