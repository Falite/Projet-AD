```python
import pandas as pd


file_path = 'resultats-par-niveau-burvot-t1-france-entiere.txt'



# Initialiser une liste pour stocker les données finales
data = []

# Initialiser une liste pour stocker les données finales
data = []

# Ouvrir le fichier en mode lecture
with open(file_path, 'r', encoding='ISO-8859-1') as file:
    # Lire la première ligne pour obtenir les noms des colonnes
    header = file.readline().strip().split(';')

    # Définir les colonnes de base et celles des candidats
    base_columns = header[:21]
    candidate_columns = ['N°Panneau', 'Sexe', 'Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp']

    # Parcourir les lignes restantes
    for line in file:
        # Diviser la ligne en valeurs en utilisant ';' comme séparateur
        values = line.strip().split(';')

        # Extraire les informations de base
        base_info = values[:21]

        # Extraire les informations des candidats
        candidate_info = values[21:]

        # Parcourir les informations des candidats en groupe de 7
        for i in range(0, len(candidate_info), 7):
            candidate_data = candidate_info[i:i+7]

            # Combiner les informations de base avec celles du candidat
            full_data = base_info + candidate_data

            # Ajouter les données complètes à la liste
            data.append(full_data)

# Créer les noms de colonnes finaux
final_columns = base_columns + candidate_columns

# Créer un DataFrame pandas avec les données transformées
df = pd.DataFrame(data, columns=final_columns)

# Afficher les premières lignes du DataFrame pour vérifier le résultat
print(df.head())

```

Standardiser les données : L'ACP est sensible à l'échelle des variables. Il est donc recommandé de standardiser les données, c'est-à-dire de les centrer réduire, pour que toutes les variables soient sur la même échelle.

```python
from sklearn.preprocessing import StandardScaler

# Sélectionner les colonnes numériques
numerical_cols = ['Inscrits', 'Abstentions', 'Votants', 'Blancs', 'Nuls', 'Exprimés', 'Voix']

# Standardiser les colonnes numériques
df_scaled = StandardScaler().fit_transform(df[numerical_cols])

# Créer un nouveau DataFrame avec les colonnes standardisées
df_standardized = pd.DataFrame(df_scaled, columns=numerical_cols)
```

Calculer les composantes principales : La méthode PCA() de sklearn permet de calculer les composantes principales à partir des données standardisées.

```python
from sklearn.decomposition import PCA

# Calculer les composantes principales
pca = PCA().fit(df_standardized)
```

Déterminer le nombre de composantes à conserver : Il existe plusieurs méthodes pour déterminer le nombre de composantes à conserver, comme la règle de Kaiser (conserver les composantes avec une valeur propre supérieure à 1) ou le coude du scree plot (rechercher le coude dans la courbe des valeurs propres).

```python
import matplotlib.pyplot as plt

# Afficher les valeurs propres
print(pca.explained_variance_)

# Afficher la variance expliquée par chaque composante
print(pca.explained_variance_ratio_)

# Tracer le scree plot
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Nombre de composantes')
plt.ylabel('Variance expliquée')
plt.show()
```

Projeter les données sur les nouvelles composantes : Une fois le nombre de composantes à conserver déterminé, il est possible de projeter les données sur les nouvelles composantes et de représenter graphiquement les individus ou les variables sur les nouveaux axes.

```python
# Conserver 2 composantes
pca = PCA(n_components=2).fit(df_standardized)

# Projeter les données sur les nouvelles composantes
df_pca = pd.DataFrame(pca.transform(df_standardized), columns=['CP1', 'CP2'])

# Ajouter les colonnes de base au DataFrame PCA
df_pca = pd.concat([df[base_columns], df_pca], axis=1)

# Représenter graphiquement les individus sur les nouveaux axes
plt.scatter(df_pca['CP1'], df_pca['CP2'])
plt.xlabel('CP1')
plt.ylabel('CP2')
plt.show()
```
